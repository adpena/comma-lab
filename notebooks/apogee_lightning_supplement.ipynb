```python
from __future__ import annotations

import json
import math
from pathlib import Path

import matplotlib.pyplot as plt

SCORE_DENOMINATOR = 37_545_489
MANIFEST_CANDIDATES = [
    Path("apogee_release_manifest.json"),
    Path("reports/graphs/site/apogee_release_manifest.json"),
    Path("reports/graphs/public_site/apogee_release_manifest.json"),
]

fallback = {
    "name": "Apogee / PR100 HNeRV-LC-v2 adapter",
    "score": 0.22826947142244708,
    "seg_dist": 0.00067623,
    "pose_dist": 0.00017198,
    "archive_bytes": 178_981,
    "archive_sha256": "afd53348f50303bf0ec6a7ffecc1ac037df2f1c70745244b9c45c72e8eb80641",
    "runtime_tree_sha256": "ef6323533666c9cac1c204a9d3f7054157d44a185b16fc859fb3f0438ccd1832",
    "samples": 600,
    "hardware": "Tesla T4 CUDA",
    "evidence_grade": "A++ exact Tesla T4 CUDA",
    "score_terms": {"SegNet": 0.067623, "PoseNet": 0.04147047142244708, "Rate": 0.119176},
    "stream_bytes": {"0.bin payload": 178_873, "zip overhead": 108},
}

def load_frontier() -> dict:
    for path in MANIFEST_CANDIDATES:
        if not path.exists():
            continue
        manifest = json.loads(path.read_text(encoding="utf-8"))
        frontier = manifest["frontier"]
        member_bytes = sum(m["bytes"] for m in frontier["archive_member_summary"]["members"])
        return {
            "name": frontier["name"],
            "score": frontier["score"],
            "seg_dist": frontier["seg_dist"],
            "pose_dist": frontier["pose_dist"],
            "archive_bytes": frontier["archive_bytes"],
            "archive_sha256": frontier["archive_sha256"],
            "runtime_tree_sha256": frontier["runtime_tree_sha256"],
            "samples": frontier["samples"],
            "hardware": frontier["hardware"],
            "evidence_grade": frontier["evidence_grade"],
            "score_terms": {"SegNet": frontier["score_terms"]["segnet"], "PoseNet": frontier["score_terms"]["posenet"], "Rate": frontier["score_terms"]["rate"]},
            "stream_bytes": {"0.bin payload": member_bytes, "zip overhead": frontier["archive_bytes"] - member_bytes},
        }
    return fallback

frontier = load_frontier()
frontier
```

```python
seg_term = 100 * frontier["seg_dist"]
pose_term = math.sqrt(10 * frontier["pose_dist"])
rate_term = 25 * frontier["archive_bytes"] / SCORE_DENOMINATOR
score = seg_term + pose_term + rate_term

print(f"name:             {frontier['name']}")
print(f"score recomputed: {score:.17f}")
print(f"reported score:   {frontier['score']:.17f}")
print(f"archive bytes:    {frontier['archive_bytes']}")
print(f"archive sha256:   {frontier['archive_sha256']}")
print(f"runtime tree:     {frontier['runtime_tree_sha256']}")
print(f"samples:          {frontier['samples']}")
print(f"hardware:         {frontier['hardware']}")
print(f"evidence:         {frontier['evidence_grade']}")
assert abs(score - frontier["score"]) < 5e-5, "score formula mismatch; check rounded component fields"
```
